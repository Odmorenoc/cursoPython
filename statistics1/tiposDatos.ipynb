### Estadisticos individualizados:

Para verificar los estadisticos de forma particular, se podrian utilizar:

1. Promedio: `(variable)[columna].mean()`
2. Mediana: `(variable)[columna].median()`
3. Moda: como la moda solo se puede realizar para datos discretos, se utiliza el histograma.
4. Histograma: es la representación grafica de las frecuencias de los datos discretos.
por lo tanto, para el caso de datos continuos, se deben generar rangos, para poder realizar el diagrama.
se realiza mediante: `(variable)[columna].plot.hist.(bins=("numero de rangos/columnas deseados"))`

```python
#Promedio
df["price_usd"].mean()
```

```python
# Mediana
df["price_usd"].median()
```

```python
#diagrama de frecuencias (histograma)
df["price_usd"].plot.hist(bins=20)
```

## Medidas de dispersión:

1. Desviación estandar: en pandas se utiliza `(Variable)["columna"].std() `
2. Rango: en pandas se utiliza `(Variable)["columna"].max() - (Variable)["columna"].min() `
3. Cuartiles = en pandas se utiliza `(Variable)["columna"].quantile(q="porcentaje que se desea") `

```python
median_price=df["price_usd"].median()
std_price= df["price_usd"].std()
q1_price=df["price_usd"].quantile(q=0.25)
q2_price=df["price_usd"].quantile(q=0.5)
q3_price=df["price_usd"].quantile(q=0.75)
max_price=df["price_usd"].max()
min_price=df["price_usd"].min()
Range_price=max_price-min_price
IQR_price=q3_price-q1_price

print(min_price, q1_price, median_price, q3_price, max_price)
print(Range_price)
```

## Limites para detección de outliners (para datos con distribución normal)

limite inferior = Q1\-1.5\*IQR

limite superior = Q3\+1.5\*IQR

```python
# ejericio de limites para una distribución normal cuando los datos estan sesgados.
minlimit=q1_price-1.5*IQR_price
maxlimit=q3_price+1.5*IQR_price

print(minlimit, maxlimit)
sns.histplot(df["price_usd"])
```
