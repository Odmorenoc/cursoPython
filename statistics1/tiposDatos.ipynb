## **Estadisticos iniciales con Pandas**

Pandas tiene un modulo que permite generar unos estadisticos iniciales de la tabla.

Se utiliza la siguiente linea de codigo: 
`(variable).describe()`

la cual nos entrega los cuartiles, moda, media, desviaci√≥n estandar, valor minimo, y valor maximo.

```python
df.describe()
```

### Estadisticos individualizados:

Para verificar los estadisticos de forma particular, se podrian utilizar:

1. Promedio: `(variable)[columna].mean()`
2. Mediana: `(variable)[columna].median()`
3. Moda: como la moda solo se puede realizar para datos discretos, se utiliza el histograma.
4. Histograma: es la representaci√≥n grafica de las frecuencias de los datos discretos.
por lo tanto, para el caso de datos continuos, se deben generar rangos, para poder realizar el diagrama.
se realiza mediante: `(variable)[columna].plot.hist.(bins=("numero de rangos/columnas deseados"))`

```python
df["price_usd"].mean()
```

```python
df["price_usd"].median()
```

```python
df["price_usd"].plot.hist(bins=20)
```
